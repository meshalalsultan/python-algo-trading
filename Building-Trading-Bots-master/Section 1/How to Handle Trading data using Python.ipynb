## Getting trading data using quandl

```python
import quandl
```

```python
aapl = quandl.get("WIKI/AAPL", start_date="2012-10-01", end_date="2017-01-01")
```

```python
type(aapl)
```

```python
aapl_numpy_array = quandl.get("WIKI/AAPL", start_date="2012-10-01", end_date="2017-01-01", returns="numpy")
```

```python
type(aapl_numpy_array)
```

```python
aapl.head()
```

```python
aapl.tail()
```

```python
fb = quandl.get("WIKI/FB", start_date="2012-01-01", end_date="2013-01-01", collapse="monthly")
```

```python
fb.head()
```

```python
fb.tail()
```
